### Face Recognition Model

This implementation of face recognition is implemented using face_recognition library using a pre-trained model.

```python
#Importing the required libraries
import face_recognition
import matplotlib.pyplot as plt

import cv2

import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
from matplotlib.patches import Circle
from scipy.spatial import distance
import numpy as np
import cv2
%matplotlib inline
```

```python
#Creating face encodings. These sample encodings are checked with the encodings of test image for a match.

# Sample encodings are of Barack Obama, Leonardo DiCaprio and George Clooney

image = cv2.imread('obama.jpg')
obama = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

image = cv2.imread('leo.jpg')
leo = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

image = cv2.imread('clooney.jpg')
clooney = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

obama_enc = face_recognition.face_encodings(obama)[0]
clooney_enc = face_recognition.face_encodings(clooney)[0]
leo_enc = face_recognition.face_encodings(leo)[0]

known_face_encodings = [
    obama_enc,
    clooney_enc,
    leo_enc
]
```

```python
#Creating a function to test images

#Function to find number of faces in the test image

def numface(imag):
    img = cv2.imread(imag)
    face_locations = face_recognition.face_locations(img)
    number_of_faces = len(face_locations)
    
    if number_of_faces==0:
        print('No faces were found in the picture!')
    if number_of_faces==1:
        print(number_of_faces,'face was found in the picture')
    elif (number_of_faces > 1):
        print(number_of_faces,'faces were found in the picture!')
    
#Function to find a match with known encodings

def find_face(img):
    image = cv2.imread(img)

    
    face_locations = face_recognition.face_locations(image)
    number_of_faces = len(face_locations)
    
    if number_of_faces ==0:
        image =  cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        plt.imshow(image)
        print("No faces found in the picture")
    else:
        unknown_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        unknown_face_encodings = face_recognition.face_encodings(unknown_image)
        face_locations = face_recognition.face_locations(image)
        image =  cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        plt.imshow(image)
        ax=plt.gca()
        for face_location in face_locations:
            top, right, bottom, left = face_location
            x,y,w,h = left, top, right, bottom
            print("The face is found at",x,y,w,h)
            rect = Rectangle((x,y), w-x , h-y, fill=False, color='red')
            ax.add_patch(rect)
        plt.show()
        for unknown_face_encoding in unknown_face_encodings:
            results = []
            for known_face_encoding in known_face_encodings:
                d = distance.euclidean(known_face_encoding, unknown_face_encoding)
                results.append(d)
            threshold =0.6
            results = np.array(results) <= threshold
            name = "Unknown"
            if results[0]:
                name = "Barack Obama"
            elif results[1]:
                name= "George Clooney"
            elif results[2]:
                name="Leonardo Di Caprio"
            if name=="Unknown":
                print('No Match')
            else:
                print(name,'was found in the picture!')
```

```python
#Testing the function with a sample image
numface("lb.jpg")
print("------------")
find_face("lb.jpg")
```
